```python
import pandas as pd
import seaborn as sns
sns.set_style("white")

# some data
sex = ['m', 'm', 'm', 'm', 'm', 'm', 'm', 'm', 'm', 'm', 'f', 'f', 'f', 'f', 'f', 'f', 'f', 'f', 'f', 'f', 'f']
time = [2.16139, 2.13694, 2.15806, 2.14972, 2.24806, 2.19278, 2.185, 2.18361, 2.13889, 2.13361, 2.42833, 2.42472, 2.45389, 2.45, 2.49167,	2.47611, 2.46833, 2.50472, 2.46861, 2.425]
df = pd.DataFrame(zip(sex, time), columns = ['sex', 'time'])

# convert category to numeric for x-axis
def cat_to_numeric(x):
    if x=='m':
        return 0
    if x=='f':
        return 1
df['sex'] = df['sex'].apply(cat_to_numeric)

# create the plot
sns.lmplot(x='sex', y='time', data=df, fit_reg=False, x_jitter = 0.02, scatter_kws={"s": 100})
sns.plt.xlabel('Sex', fontsize=18)
sns.plt.ylabel('Time (hrs)', fontsize=18)
sns.plt.xticks((0,1), ('Male', 'Female'), fontsize = 12)
sns.plt.show()
```
