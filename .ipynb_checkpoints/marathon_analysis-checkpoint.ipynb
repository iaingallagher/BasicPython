## Analysis of Marathon Times

In this tutorial we're going to use python to analyse marathon winning times. These data are available from [OpenIntro Statistics](http://www.openintro.org/stat/). We'll use the ```pandas``` library to hold and manipulate our data, ```pyplot``` from the ```matplotlib``` library (often via ```pandas```) to plot the data and the ```scipy``` library for statistical analyses. First we import the libraries that we'll be using.

## Preliminaries

````python
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats as stats
````

## Get the data

The data are a set of marathon winning times for the years 1970 to 2000 for men and women. These data are held in a ```csv``` file and we'll read that into a ```pandas``` DataFrame object using the [```read_csv```](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.io.parsers.read_csv.html) function. Then we'll examine the first 5 rows of the data to make sure it looks ok using the ```.head``` method on our DataFrame.

````python
data_in = pd.read_csv('marathon.csv', sep=',')
data_in.head()
````

We can get information about the DataFrame structure using the ```.info()``` method on our DataFrame. This tells us we have 59 rows and 3 variables. ```Year``` is an integer, ```Gender``` is an object (here a string) and ```Time``` is an integer. Note that ```Gender``` is reported as an ```object``` rather than a ```string``` because of the way ```pandas``` works internally.

````python
data_in.info()
````

````python
data_in.dtypes
````

The data types returned by ```pandas``` are based on those of ```numpy```. The ```object``` datatype is returned for strings because the ```ndarray``` holding the strings has the ```object``` datatype (as opposed to ```int64``` for example). We can look at the datatypes more explicitly by looking at the type of the first entry in each column (```ix.[0]``` below).

````python
for i in data_in.columns:
    print "Column %s is %s." % (i, type(data_in[i].ix[0]))
````

One of the first questions we might have about these data is 'what is the mean winning time?'. We can calculate the mean quite simply with ```pandas```.

````python
mean_time = data_in['Time'].mean()
# data_in.Time.mean() alternative syntax
print "The mean marathon winning time is %.2f hours." % mean_time 
````

So now we know the mean winning time but what about the time for each gender? Are these different? We can subset the data by gender and then calculate the mean time.

````python
male_mean_time = data_in[data_in.Gender == 'm'].Time.mean()
female_mean_time = data_in[data_in.Gender == 'f'].Time.mean()

print "The average winning marathon time for men is %.2f hours and the average winning time for women is %.2f hours." % (male_mean_time, female_mean_time)
````

So we can see that male winning times tend to be ~0.3 hours (or 20 minutes) faster than female winning times over the 30 year span of the data.

We could look at this graphically by plotting the winning times by gender using [boxplots](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.boxplot.html). We can do this by calling ```matplotlib``` via ```pandas``` using the ```boxplot``` method on our dataframe.

````python
data_in.boxplot(column='Time', by='Gender', grid=False) # generates plot but not yet shown
plt.title('Time by Gender') # add title and axis labels
plt.xlabel('Gender')
plt.ylabel('Time')
plt.suptitle('') # get rid of automatic title
plt.show() # show the plot
````

From the plot it's obvious that the male and female times are quite different. We probably want to know if thet're significantly different. First we want to check the 'normality' of the data. One way to do that is to examine how normal a [histogram](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.hist.html) of the data looks.

````python
data_in.hist(column='Time', by=data_in.Gender, grid=False, bins=10)
plt.show()
````

From this it's easy to see that neither of the distributions of the time variable don't look particularly normal. In addition to histograms we can also examine the normality of the data with quantile-quantile plots. In these plots the ordered data are plotted against the quantiles of the standard normal distribution. If the data are normal the plot should be a straight line. To create a q-q plot we have to use the ```probplot``` function from the ```scipy.stats``` module. We imported this module as ```stats``` at the beginning of our analysis.

````python
male_times = data_in[data_in.Gender=='m'].Time # get separate time for each gender
female_times = data_in[data_in.Gender=='f'].Time

figure(figsize(10,6)) # set a figure size so plots show up

subplot(121) # draw first plot
stats.probplot(male_times, dist="norm", plot=pylab)
plt.title('Male Data Q-Q Plot')

subplot(122) # draw 2nd plot
stats.probplot(female_times, dist="norm", plot=pylab)
plt.title('Female Data Q-Q Plot')

pylab.show()
````

As in the histograms we can see that these data are not normal and there are a few years with slow winning times in both the male and female data that drag the distribution away from normal. One test to examine the difference between the male an female times is the $t$-test. This is a parametric test and relies on the data being approximately normal. However the $t$-test is pretty robust to non-normality especially when we have n>30 data points.

````python
data_in.Gender.value_counts()
````

We can see that we have 30 male observations and 29 female observations so we're probably ok as far as non-normality and $t$-testing goes. We'll use the [```scipy.stats.ttest_ind```](http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html#scipy.stats.ttest_ind) for this and pass in the ```male_times``` and ```female_times``` we defined above when we created our qqplots. The ```stats.ttest_ind``` module returns a tuple containing the t-statistic and the p-value for the test.

````python
t_res = stats.ttest_ind(male_times, female_times) # tuple, t-stat and p-value
print "The t-statistic is %.2f with a p-value of %.3f." % (t_res[0], t_res[1])
````

The $t$-statistic absolute value is >2 and the $p$-value is very low (< 0.001). Both of these indicate that there is a statistically significant difference between the male and female marathon winning times. However we should be concerned about the non-normality of the data. We can deal with this to by using the non-parametric equivalent of the $t$-test - the Mann-Whitney U test (or Wilcoxon rank sum test). We carry this test out using ```ranksums``` from the ```scipy.stats```.

````python
mw_test = stats.ranksums(male_times, female_times) # tuple, z-value and p-value
print "The p-value for the MW test of winning times by gender is %.3f." % mw_test[1]
````

Like the parametric test this suggests the male and female winning times are significantly different.

## Winning time trends

We might also want to investigate the trend in winning times over the years using this data. The first thing we want to do is plot the data. We have two numerical variables (Time and Year) and so we want a scatterplot. First we set up different colours for the points depending on whether the times are from the male or female data. In this case if the data point is not male use red and if it is male use blue. See the [numpy.where](http://docs.scipy.org/doc/numpy/reference/generated/numpy.where.html) documentation for more information.

````python
colors = np.where(data_in.Gender!='m', 'r', 'b')
plt.scatter(data_in.Year, data_in.Time, c=colors)
plt.title('Time by Year')
plt.xlabel('Year')
plt.ylabel('Time')
plt.show()
````

We can test the correlation between the Year and Time using the ```pearsonr``` test from the ```scipy.stats``` module. Like the ```ttest_ind``` module above this returns a tuple but in this case the values are Pearson's $\rho$ and the $p$-value for the correlation between the datasets.

````python
male_corr = stats.pearsonr(data_in.Year[data_in.Gender=='m'], male_times)
female_corr = stats.pearsonr(data_in.Year[data_in.Gender=='f'], female_times)
print "The correlation between Year and Time for the male data is %.2f with a p-value of %.4f" % male_corr
print "The correlation between Year and Time for the female data is %.2f with a p-value of %.4f" % female_corr
````

The correlation between Year and Time is highly significant and in both cases Pearsons $\rho$ is negative so times are decreasing over the years. It would be nice to be able to visualise this on the plot by adding the best fit line for each dataset. First we get the coefficients for the lines of best fit by fitting a first degree polynomial to the data. Then we plot the data and add a line showing the best fit.

````python
male_m, male_b = np.polyfit(data_in.Year[data_in.Gender=='m'], male_times, 1) # fit polynomial of degree 1 - that last 1 parameter
female_m, female_b = np.polyfit(data_in.Year[data_in.Gender=='f'], female_times, 1)

# Just like the plot above but add regression lines.
colors = np.where(data_in.Gender!='m', 'r', 'b')
plt.scatter(data_in.Year, data_in.Time, c=colors) # create plot
plt.plot(data_in.Year[data_in.Gender=='m'], data_in.Year[data_in.Gender=='m']*male_m+male_b, '-', color='b', label='Male') # add line to plot
plt.plot(data_in.Year[data_in.Gender=='f'], data_in.Year[data_in.Gender=='f']*female_m+female_b, '-', color='r', label='Female')
plt.title('Winning Time by Year')
plt.legend(loc='upper right')
plt.xlabel('Year')
plt.ylabel('Time')

plt.show()
````

The lines fitted to both sets of data indicate that there is a negative relationship between year the race is run and the winning time - so winning times have become lower over the years. If we look at these plots we see that since about 1977 the winning times have varied very little. So we might want to examine that more closely by re-calculating the correlation and re-drawing the above plot for data since 1977.

````python
recent_subset = data_in[data_in.Year > 1977] # filter the data for recent times only
recent_male_corr = stats.pearsonr(recent_subset.Year[recent_subset.Gender=='m'], recent_subset.Time[recent_subset.Gender=='m'])
recent_female_corr = stats.pearsonr(recent_subset.Year[recent_subset.Gender=='f'], recent_subset.Time[recent_subset.Gender=='f'])

print "The correlation between Year and Time for the recent male data is %.2f with a p-value of %.4f" % recent_male_corr
print "The correlation between Year and Time for the recent female data is %.2f with a p-value of %.4f" % recent_female_corr
````

Neither of these correlations are significant although the Pearson's $\rho$ for each is still negative. Finally we'll plot this more recent data.

````python
recent_male_m, recent_male_b = np.polyfit(recent_subset.Year[data_in.Gender=='m'], recent_subset.Time[recent_subset.Gender=='m'], 1) # fit polynomial of degree 1 - that last 1 parameter
recent_female_m, recent_female_b = np.polyfit(recent_subset.Year[recent_subset.Gender=='f'], recent_subset.Time[recent_subset.Gender=='f'], 1)
# Just like the plot above but add regression lines.
colors = np.where(recent_subset.Gender!='m', 'r', 'b')
plt.scatter(recent_subset.Year, recent_subset.Time, c=colors)
plt.plot(recent_subset.Year[recent_subset.Gender=='m'], recent_subset.Year[recent_subset.Gender=='m']*recent_male_m+recent_male_b, '-', color='b', label='Male') # add line to plot
plt.plot(recent_subset.Year[recent_subset.Gender=='f'], recent_subset.Year[recent_subset.Gender=='f']*recent_female_m+recent_female_b, '-', color='r', label='Female') # add line to plot
plt.title('Recent Winning Times')
plt.legend(loc='upper right')
plt.xlabel('Year')
plt.ylabel('Time')

plt.show()
````

These lines look considerably flatter than the lines shown above.
