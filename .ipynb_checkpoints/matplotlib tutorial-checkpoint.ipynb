We'll begin by simply drawing the ```sine``` and ```cosine``` functions in a simple line plot and then progressively tidy up and improve the plot.

````python
X=np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
````

So now we have two variables ```C``` and ```S``` containing 256 values of $cos(X)$ and $sin(X)$ respectively from $-\pi$ to $\pi$. Now we'll create the plots and ```show()``` them.

````python
plot(X,C)
plot(X,S)
show()
````

This plot is fine for exploration perhaps but we can control many of the parameters such as figure size, line styles, colours, axes and axis and label properties etc. Let's begin by changing the figure size and plotting at a different resoultion (in this case 80dpi).

````python
figure(figsize=(5,3), dpi=80)
plot(X,C)
plot(X,S)
show()
````

Now let's change some properties of the lines used to plot each function of ```X```. The ```linewidth``` below is set in pixels.

````python
plot(X,C, color='red', linewidth=1.0, linestyle='-')
plot(X,S, color='blue', linewidth=2.0, linestyle='--')
show()
````

Now we'll set the axis limits to be between -3 and 3 to make the x-axis tighter.

````python
plot(X,C)
plot(X,S)
xlim(-3,3)
show()
````

Now we'll use ```numpy``` (imported as ```np``` when Ipython notebook starts) to create x-axis ticks that reflect the values we care about (i.e. $-\pi$ and $\pi$. Notice we give the ```xticks``` and ```yticks``` functions a list of values where we'd like tick marks placed.

````python
plot(X,C)
plot(X,S)
xlim(-4,4)
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
yticks([-1,0,1])
show()
````

So that's better but it would be even better if the values on the x-axis were expressed as $\pi$ or fractions of $\pi$ instead of simply 3.14 etc. We can do this by providing Latex labels in a second list to the ```xticks``` and ```yticks``` arguments.

````python
plot(X,C)
plot(X,S)
xlim(-4,4)
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
yticks([-1,0,1])
show()
````

Now we'll change the positions of the plot spines. The spines are the lines which demarcate the plot boundary above. In addition they connect the plot tick marks. We'll set the spines so they meet at the origin of the plot and also move the tick marks from above the spines to below in the x-axis.

````python
plot(X,C)
plot(X,S)
xlim(-4,4)

# gca() means get current axes
ax = gca()
ax.spines['right'].set_visible(False) # lose the right hand spine
ax.spines['top'].set_visible(False) # lose the top spine

ax.xaxis.set_ticks_position('bottom') # show ticks on bottom spine
ax.spines['bottom'].set_position(('data', 0)) # pass a tuple to set position
ax.spines["bottom"].axis.axes.tick_params(direction='outward') # set tickmark direction

ax.yaxis.set_ticks_position('left') # show ticks on left spine
ax.spines['left'].set_position(('data', 0))

xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
yticks([-1,0,1]) # now re-label

show()
````
