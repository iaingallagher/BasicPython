# Font effects

You can make text *italic* using a single asterisk (*) before and after the word/text.
You can make text **bold** by using a double asterisk before and after the word/text.
You can enter ```monotype``` text (i.e. looks like code) by using three backticks before and after the word/text.

# Embedding links

You can embed links.

# Code in jupyter notebook

Create a new cell. The cells are automatically created as code cells. Enter some code. Press the play button to execute that code.

````python
# import libraries we need
import pandas as pd
import matplotlib.pyplot as plt
# so we can see plots in notebook
% matplotlib inline 

# get the data into pandas
data = pd.ExcelFile('data/obesity_data.xlsx')

# just get the data we want
# Read 2nd section, the obesity figures by age data
data_age = data.parse('7.2', skiprows=3, skipfooter=14) # skip first 3 rows, last 14 rows
data_age.head() # first few entries
````

As we can see there are NA values in the data. We cannot analyse these so we will drop the rows and columns containing NA values from the dataset.

````python
# drop NA and check
data_age.dropna(axis = (0,1), how='all', inplace=True) # drop NaN
data_age.head()
````
