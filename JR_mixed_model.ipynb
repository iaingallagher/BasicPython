```python
# based on http://nbviewer.jupyter.org/urls/umich.box.com/shared/static/6tfc1e0q6jincsv5pgfa.ipynb

import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

data_in = pd.read_csv('data/BODY_COMPOSITION_DATA_4_ASSIGNMENT_INDIV.csv', sep=',')
data_in.head()
```

```python
data_in.dropna(axis = (0,1), how='any', inplace=True) # drop NaN
data_in.head()
data_in.describe()
```

```python
data_inM = pd.melt(data_in, id_vars=['indiv'], value_vars=['girths', 'bia', 'DW', 'jackson', 'HW'])
data_inM.head()
```

```python
md = smf.mixedlm("value ~ indiv", data_inM, groups=data_inM["variable"])
mdf = md.fit()
print(mdf.summary())
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd as tukeyhsd
post_hoc = tukeyhsd(data_inM['value'], data_inM['variable'], 0.05) # apply tukey HSD
print(post_hoc)
```
